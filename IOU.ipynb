```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
# IOU (Intersection Over Union)

def iou (img, target_xy, target_width, target_height, guess_xy, guess_width, guess_height) : # target, guess : (x, y, w, h)
    
    img = plt.imread(img)
    plt.imshow(img)
    ax = plt.gca()
    
    target_rect = Rectangle(target_xy, target_width, target_height, linewidth=1,edgecolor = 'green', facecolor = 'none')
    guess_rect = Rectangle(guess_xy, guess_width, guess_height, linewidth=1, edgecolor = 'red',facecolor = 'none')
    
    intersect = intersection(target_xy, target_width, target_height, guess_xy, guess_width, guess_height)
    
    intersection_rect = Rectangle(intersect[:2], intersect[2], intersect[3], facecolor = 'yellow')
    
    ax.add_patch(target_rect)
    ax.add_patch(guess_rect)
    ax.add_patch(intersection_rect)
    
    
def intersection (xy1, width1, height1, xy2, width2, height2) :
    x_start1, y_start1 = xy1 
    x_end1, y_end1 = ( x_start1+width1, y_start1 + height1 )
    
    x_start2, y_start2 = xy2
    x_end2, y_end2 = ( x_start2+ width2, y_start2 + height2 )
    
    x_max_start = max(x_start1, x_start2)
    y_max_start = max(y_start1, y_start2)
    
    x_min_end = min(x_end1, x_end2)
    y_min_end = min(y_end1, y_end2)
    
    return ( x_max_start, y_max_start, x_min_end - x_max_start, y_min_end - y_max_start ) # (x,y,w,h)
    
```

```python
imgpath = 'Downloads/naro.jpeg'

img = plt.imread(imgpath)

print(img.shape)

plt.imshow(img)

ax = plt.gca()
guess_xy = (0, 350)
guess_width = 220
guess_height = 250

target_xy = (0, 300)
target_width = 250
target_height = 220
```

```python
iou(imgpath, target_xy, target_width, target_height, guess_xy, guess_width, guess_height)

```
