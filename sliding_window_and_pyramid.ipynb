```python
import cv2
import matplotlib.pyplot as plt
from glob import glob
```

```python
def sliding_window(img, step, ws) :
    
    for y in range(0, img.shape[0] - ws[1], step) :
        for x in range(0, img.shape[1] - ws[0], step) :
            yield (x,y, img[y:y+ws[1], x:x+ws[0]])

            
test_imgs = glob('../../Downloads/cat_location/*.jpg')
samp = cv2.imread(test_imgs[0])
plt.imshow(samp)
 
print(samp.shape)
    
cnt = 0
            
for (x,y,img) in sliding_window(samp, 50, (30,30)) :
    
    cnt +=1 
    plt.imshow(img)
    plt.show()

print(cnt)
```

```python
def pyramid(img, scale, min_size) : # ex (img, 1.5, min_size = (100, 100))
    # yield original image
    yield img
    
    while True :
        
        img = cv2.resize(img, (int(img.shape[1]/scale), int(img.shape[0]/scale)))
        
        if img.shape[1] < min_size[0] or img.shape[0] < min_size[1] :
            break
        
        yield img
           
        
for img in pyramid(samp, 2, (50, 50)) :
    plt.imshow(img)
    plt.show()
```
